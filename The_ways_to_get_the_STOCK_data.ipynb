API Methods:
returnChartData

```python
import requests
import json
import pandas as pd



# BITCOIN
req= requests.get('https://poloniex.com/public?command=returnChartData&currencyPair=BTC_XMR&start=1405699200&end=9999999999&period=14400')
data= json.loads(req.text)

df=pd.DataFrame(data)
df
```
