## EXPECTATION - MAXIMIZATION and Converge

Maximum Likelihood or  EM Algorithm is to find the relevant parameters in the the presence of the latent variables. A set of starting parameters in the imcomplete data after that comes the EXPECTATION step ( Estimation of the missing values in data.) and guessing the the value in the missing data after the MAXIMILIZATION step( updates the missing values in data and Last step we execute the steps EXPECTATION and MAXIMIZATION till The convergence is met.

Convergence ( yakinsamak ) in probability is based on intuition. For example , we have 2 random variables and if the probability of their difference is very small. It is called converged ( birbirine yaklastirildi ).Conversion means if the values match each other .

1-If we have missing values

2- put some values that matches our **expectation** ( expecting the imcomplete value or data)

3- **maximization** will generate the data after completion

4- we will check if the missing value is matching the expected value or not.

If the conversion is not met , we will repeat **Expectation** and maximization steps until the conversion is met.( corected value )

## Gaussian Mixture Models

The GMM or Gaussian Mixture Model is a mixture model that uses a combination of probability distributions and requires the estimation of mean and standard deviation parameters.

```python
from numpy import hstack
import matplotlib.pyplot as plt
from numpy.random import normal
```


numpy.hstack() function is used to stack the sequence of input arrays horizontally (i.e. column wise) to make a single array.

```python
sample1= normal ( loc=20, scale =5 , size =4000)
sample2= normal ( loc=40, scale =5 , size =8000)
```

```python
sample = hstack((sample1,sample2))
```

```python
plt.hist(sample, bins=50 , density=True , color='red')
```

```python
from sklearn.mixture import GaussianMixture
```

```python
GMsample= sample.reshape((len(sample),1))
```

```python
model=GaussianMixture(n_components=3, init_params= "random", covariance_type='full')
```

```python
model.fit(GMsample)
```

```python
y1=model.predict(GMsample)
```

```python
print(y1[:100])
print(y1[-100:])
```
